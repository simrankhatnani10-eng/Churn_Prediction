### Data Cleaning
- Remove Unneccessary Columns
- Data type Mismatch

```python
import sys
from pathlib import Path
PROJECT_ROOT = Path().resolve().parent
sys.path.append(str(PROJECT_ROOT))
```

```python
# Import Modules
import numpy as np
import pandas as pd
from SRC import config
```

```python
# Load data
DATA_FILE_PATH = config.DATA_PATH
CLEANED_DATA_FILE_PATH = config.CLEANED_DATA_PATH
print("Raw Data path: \n", DATA_FILE_PATH)
print("Cleaned Data Path: \n", CLEANED_DATA_FILE_PATH)
```

```python
# load data
data = pd.read_csv(DATA_FILE_PATH)
data.head()
```

```python
# Back-up
df = data.copy()
```

```python
# Meta data

print("No of Rows: ",df.shape[0])
print("\nNo of columns: ",df.shape[1])
print("\nAll columns: \n",df.columns.to_list())
print("Data types: ",df.dtypes)
print("/n info")
df.info()
```

```python
# Change the data type
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors = "coerce")
print("Datatype of 'Total Charges': ",df["TotalCharges"].dtypes)
```

```python
# Remove Irrelevant columns
df.drop(columns = ["customerID"] , inplace = True)
print("Final list of columns: ",df.columns.to_list())
```

```python
# save cleaned data
df.to_csv(CLEANED_DATA_FILE_PATH, index = False)
print("Data is saved successfully")
```
